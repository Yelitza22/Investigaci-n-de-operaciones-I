```python
from sympy import plot_implicit, symbols, Eq, And
```

```python
x,y=symbols('x,y')
```

```python
p1=plot_implicit(And(x<5-2*y,y<4-2*x),(x,0,6),(y,0,5))
p2=plot_implicit(Eq(2*x+3*y,5),(x,0,6),(y,0,5))
```

```python
p2.append(p1[0])
p2[0].line_color='orange'
p2.show()
```

```python
p3=plot_implicit(And(x<6-3*y,y<3-(3/2)*x),(x,0,6),(y,0,5))
p4=plot_implicit(Eq(2*x+3*y,5),(x,0,6),(y,0,5))
```

```python
p3.append(p4[0])
p3[0].line_color='orange'
p3.show()
```
