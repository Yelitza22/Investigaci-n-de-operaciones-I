Trabajeremos con sympy y haremos ejemplos de las gráficas para mostrar las ventajas de usar dicha librería


```python
import sympy as sp #La libreria es para cálculo simbólico tipo wolfrand
sp.init_printing()
```

```python
x=sp.symbols('x')
y=sp.Function('y')
```

```python
ode =sp.Eq(y(x)*sp.Derivative(y(x),x)+2*x,0)
```

```python
ode
```

```python
sp.classify_ode(ode) #Esta describiendo la ecuación
```

```python
sol=sp.dsolve(ode,y(x)) #solución de la ecuación dada
```

```python
sol #Para ver la solución
```

```python
c1=sp.plotting.plot(x**6) #en esta librería podemos graficar más rápido, solo dandole la ecuacion
```

```python
c1=sp.plotting.plot(x**6,show=False) #no lo muestra pq dice False=no lo muestre
```

```python
c1=sp.plotting.plot(x*x,show=True) #Con True se muestra
c2=sp.plotting.plot(x**3,show=True) #si ponemos ambas las imprime separadas aunque esten en la misma celda
```

```python
c2 #Nos dice el espacio de memoria donde esta la gráfica
```

```python
c2.show() #Con el show hace que la muestre
```

```python
c2.append(c1[0]) #anida las gráficas la c1 sobre la c2, el [0] es para indicar que es un arreglo
c2.show() #para que las muestre
```

```python
c1.extend(c2) #anida las graficas
c1.show()
```

```python
from sympy.plotting import plot3d #importa el módulo de gráficar en 3d
```

```python
x,y=sp.symbols('x,y') #Cada que traemos un módulo, tenemos que llamar a las varaiables para su ejecución
```

```python
plot3d(x*y, (x,-1,1), (y,-1,1)) #Para ver la gráfica (x,-1,1) y (y,-1,1) es para acotar los ejes
```

```python
plot3d(x*x+y**2, (x,-10,10), (y,-10,10)) #Gráfica de x^2+y^2
```

```python
plot3d((x**2+y**2, (x,-10,10), (y,-10,10)), ((x-y),(x,-10,10),(y,-10,10))) #Para graficar dos funciones a la vez, tener cuidado en los parentesis
```

```python
from sympy.plotting import plot3d_parametric_surface #Grafica de superficie en 3d
```

```python
u,v=sp.symbols('u,v')
```

```python
plot3d_parametric_surface(sp.cos(u+v),sp.sin(u-v),u-v)
```

```python
plot3d_parametric_surface(sp.cos(u)*sp.cos(v),sp.sin(v)*sp.cos(u),u-v)
```

```python
from sympy import plot_implicit, symbols, Eq, And #Pedimos de la libreria las 4 funciones a ocupar
```

```python
x,y=symbols('x,y') #Definimos las variables
```

```python
p1=plot_implicit(Eq(x**2+y**2,5)) #lado izaquierdo, lado derecho de la igualdad
```

```python
p2=plot_implicit(y>x**2)
```
