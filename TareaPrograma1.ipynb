## Ejercicio 1
Una compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor A gana \$3 por motor y con el motor B gana \$4 por motor.
Dispone de 80 horas por semana para producirlos y en los procesos se requieren: 4 h para A y solo 2 h para B. Sin embargo, debido a lo peligroso del material usado, son necesarias 5 h de preparación y limpieza para el motor B y 2 h para A. El tiempo de preparación total disponible es 120 h por semana. Determinar el número de cada clase de motores a producir que dé como resultado la máxima ganancia.

## Variables de decisión:

*   x = Número de motores A a producir
*   y = Número de motores B a producir

##Función Objetivo:

Maximizar   z = 3x + 4y

##Restricciones:

Consideramos las horas de produción por semana y las horas de preparación por semana, así:

*   R1: 4x + 2y <= 80
*   R2: 2x + 5y <= 120

Con x,y valores reales positivos




```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()

# Restricción para R1
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40, 0], color='blue', alpha=0.25, label='Resultante de R1')

plt.xlabel('Número de motores A a producir')
plt.ylabel('Número de motores B a producir')
plt.legend()
plt.show()
```
